```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def max_diagonal_element(matrix):
    max_element = matrix[0][0] 

    for i in range(len(matrix)):
        if matrix[i][i] > max_element:
            max_element = matrix[i][i]

    return max_element
```

```python
def Levenberg_Marquart(R,J,z0,N,mu0,tol,x,y):
    res = 0
    n = len(z0)
    Id = np.eye(n)
    R0 = R(z0[0],z0[1],z0[2],x,y)
    J0 = J(z0[0],z0[1],z0[2],x,y)
    f0 = 0.5*R0.T@R0
    A = J0.T@J0
    g = J0.T@R0
    md = max_diagonal_element(A)
    mu = min(mu0,md)
    
    k = 0
    while(k < N+1):
        pk = np.linalg.solve(A + mu*Id, -g)
        if np.linalg.norm(pk) < tol:
            res = 1
            break
        z0 = z0 + pk
        R1 = R(z0[0],z0[1],z0[2],x,y)
        f1 = 0.5*R1.T@R1
        
        ro = (f0-f1)/(-pk.T@J0.T@R0 + 0.5*mu*pk.T@pk)
        
        if ro < 0.25:
            mu = 2.0*mu
        if ro > 0.75:
            mu = mu/3.0
            
        J0 = J(z0[0],z0[1],z0[2],x,y)
        A = J0.T@J0
        g = J0.T@R1
        f0 = f1
        R0 = R1
        
        k = k + 1 
        
    return z0,f1,pk,k,res
```

```python
def R(A,w,phi,x,y):
    n = x.shape[0]
    vec = np.zeros(n)
    for i in range(n):
        vec[i] = A*np.sin(w*x[i] + phi) - y[i] 
        
    return vec
```

```python
def J(A,w,phi,x,y):
    n = x.shape[0]
    mat = np.zeros((n,3))
    
    for i in range(n):
        mat[i,0] = np.sin(w*x[i] + phi)
        mat[i,1] = A*np.cos(w*x[i] + phi)*x[i]
        mat[i,2] = A*np.cos(w*x[i] + phi)
        
    return mat
```

```python
#Cargamos los datos
datos = np.load('puntos2D_ej3.npy')
x = datos[:,0]
y = datos[:,1]
```

### Primer experimento

```python
#Inicializamos las variables
z0 = np.array([15.0,0.6,0.0])
N = 5000
mu0 = 0.001
eps = np.finfo(np.float64).eps
tol = np.sqrt(eps)
```

```python
#Vamos a imprimir el valor inicial
print(0.5*R(z0[0],z0[1],z0[2],x,y).T@R(z0[0],z0[1],z0[2],x,y))
```

```python
z1,f1,pk,k,res = Levenberg_Marquart(R,J,z0,N,mu0,tol,x,y)

if res == 1:
    print("El algoritmo convergió")
else:
    print("El algoritmo no convergió")
    
    
print("zk = ", z1)
print("f(zk) = ", f1)
print("||pk||", np.linalg.norm(pk))
print("Número de iteraciones: ", k)
```

```python
#Calculamos los extremos del intervalo a graficar
min_x = min(x)
max_x = max(x)

n = x.shape[0]

w = np.linspace(min_x, max_x, 100)
z = z1[0]*np.sin(z1[1]*w + z1[2])
```

```python
plt.plot(x, y, label='Puntos de la muestra')
plt.plot(w, z, label='Modelo')
plt.legend()
plt.show()
```
